# **Pivot Basics**
Pivot is used to transform or reshape data

```python
import pandas as pd
import numpy as np
df = pd.read_csv("weather.csv")
df
```

```python
df.pivot(index='city', columns='date')
#city column turns into index column
#the whole date column turns into a single row acting as individual columns
```

```python
df.pivot(index='city', columns='date', values='humidity')
#when values argument is used only those values will be printed otherwise everything will be printed
```

```python
df.pivot(index='date', columns='city')
```

```python
df.pivot(index='humidity', columns='city')
```

# Pivot Table

```python
df = pd.read_csv("weather2.csv")
df
```

```python
df.pivot_table(index='city', columns='date')
```

# Margins

```python
df.pivot_table(index="city",columns="date", margins=True,aggfunc=np.sum)
#margins=true creates a all column n row n shows the sum
#aggfunc=np.sum adds the similar date values and respective city values
#aggfunc can be used for count,mean and stdev
```

# Grouper

```python
df = pd.read_csv("weather3.csv")
df
```

```python
df['date'] = pd.to_datetime(df['date']) 
#converting scalar dates to pandas datetime object
```

```python
df.pivot_table(index=pd.Grouper(freq='M', key='date'), columns='city')
#setting frequency to month
#output is the average value of that particular month
```

# Crosstabs

```python
df4 = pd.read_excel("survey.xlsx")
df4
```

```python
pd.crosstab(df4.Nationality,df4.Handedness)
```

```python
pd.crosstab(df4.Sex,df4.Handedness)
```

# Margins

```python
pd.crosstab(df4.Sex,df4.Handedness, margins=True)
```

# Multi Index Column and Rows

```python
pd.crosstab(df4.Sex, [df4.Handedness,df4.Nationality], margins=True)
```

```python
pd.crosstab([df4.Nationality, df4.Sex], [df4.Handedness], margins=True)
```

# Normalise

```python
pd.crosstab(df4.Sex, df4.Handedness, normalize='index')
```

## Aggfunc and Values

```python
import numpy as np
pd.crosstab(df4.Sex, df4.Handedness, values=df4.Age, aggfunc=np.average)
```

#Testing for strings that match or contain a pattern

```python
pattern = r"[0-9][a-z]"
pd.Series(["1", "2", "3a", "3b", "03c", "4dx"],dtype="string",).str.contains(pattern)
```

```python
pd.Series(
    ["1", "2", "3a", "3b", "03c", "4dx"],
    dtype="string",
).str.match(pattern)
```

```python
pd.Series(
    ["1", "2", "3a", "3b", "03c", "4dx"],
    dtype="string",
).str.fullmatch(pattern)
```
