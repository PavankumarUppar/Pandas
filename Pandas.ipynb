# **Pivot Basics**

```python
import pandas as pd
import numpy as np
df = pd.read_csv("weather.csv")
df
```

```python
df.pivot(index='city', columns='date')
```

```python
df.pivot(index='date', columns='city')
```

# Pivot Table

# Crosstabs

```python
df4 = pd.read_excel("survey.xlsx")
df4
```

```python
pd.crosstab(df4.Nationality,df4.Handedness)
```

```python
pd.crosstab(df4.Sex,df4.Handedness)
```

# Margins

```python
pd.crosstab(df4.Sex,df4.Handedness, margins=True)
```

# Multi Index Column and Rows

```python
pd.crosstab(df4.Sex, [df4.Handedness,df4.Nationality], margins=True)
```

```python
pd.crosstab([df4.Nationality, df4.Sex], [df4.Handedness], margins=True)
```

# Normalise

```python
pd.crosstab(df4.Sex, df4.Handedness, normalize='index')
```

## Aggfunc and Values

```python
import numpy as np
pd.crosstab(df4.Sex, df4.Handedness, values=df4.Age, aggfunc=np.average)
```

# Working with text data

1. object -dtype NumPy array.

2. StringDtype extension type.

```python
pd.Series(["a", "b", "c"])
```

```python
pd.Series(["a", "b", "c"], dtype="string")
```

```python
s = pd.Series(["a", "b", "c"])
s
```
